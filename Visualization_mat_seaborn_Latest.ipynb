# seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
tips =sb.load_dataset('tips')
```

```python
tips
```

## Univariate  one variable data visualization

```python
#strip plot
sb.stripplot(y= 'tip', data=tips,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Grouping with stripplot()
sb.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sb.stripplot(x='day', y='tip', data=tips, size=4,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarm plot
sb.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
    
```

```python
sb.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
#Changing orientation
sb.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

### Box and Violin plot

```python

sb.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
```

```python

sb.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
plt.subplot(1,2,1)
sb.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.subplot(1,2,2)
sb.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
##Combining plots

sb.violinplot(x='day', y='tip', data=tips, inner=None,color='lightgray')
sb.stripplot(x='day', y='tip', data=tips, size=4,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

## Bivariate  two variables data visualization

```python
#Joint plot

sb.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
# Density plot
sb.jointplot(x='total_bill', y= 'tip', data=tips,kind='kde')
plt.show()
```

```python
#Pair plot

sb.pairplot(tips)
plt.show()
```

```python
sb.pairplot(tips, hue='sex')
plt.show()
```
