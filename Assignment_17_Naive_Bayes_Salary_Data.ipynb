```python
import pandas as pd
import numpy as np
sal_train = pd.read_csv("SalaryData_Train.csv")
sal_test =pd.read_csv("SalaryData_Test.csv")
```

```python
string_columns = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
```

```python
from sklearn.preprocessing import LabelEncoder
number = LabelEncoder()
```

```python
for i in string_columns:
    sal_train[i]= number.fit_transform(sal_train[i])
    sal_test[i]=number.fit_transform(sal_test[i])
```

```python
sal_train
```

```python
sal_train.shape
```

```python
sal_train.describe()
```

```python
x_train = sal_train.iloc[:,0:13]
y_train = sal_train.iloc[:,13]
x_test = sal_test.iloc[:,0:13]
y_test = sal_test.iloc[:,13]
```

```python
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB
```

```python
# Multinomial Naive Bayes
classifier_mb = MB()
classifier_mb.fit(x_train, y_train)
train_pred_m = classifier_mb.predict(x_train)
accuracy_train_m = np.mean(train_pred_m == y_train) # 77%

test_pred_m = classifier_mb.predict(x_test)
accuracy_test_m = np.mean(test_pred_m==y_test) # 77%
```

```python
accuracy_test_m
```

```python
pd.crosstab(test_pred_m,y_test)
```

```python
# Gaussian Naive Bayes 
classifier_gb = GB()
classifier_gb.fit(x_train, y_train) # we need to convert tfidf into array format which is compatible for gaussian naive bayes
train_pred_g = classifier_gb.predict(x_train)
accuracy_train_g = np.mean(train_pred_g == y_train) #80%

test_pred_g = classifier_gb.predict(x_test)
accuracy_test_g = np.mean(test_pred_g==y_test) # 80%
```

```python
accuracy_test_g
```

```python
pd.crosstab(test_pred_g,y_test)
```

In comparision GaussianNB has the best Accuracy
