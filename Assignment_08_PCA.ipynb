```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
wine = pd.read_csv("wine.csv")
wine.describe()
wine.head()
```

```python
# Considering only numerical data 
wine.data = wine.iloc[:,1:]
wine.data.head()
```

```python
# Converting into numpy array
wine1 = wine.data.values
wine1
```

```python
# Normalizing the numerical data 
wine_normal = scale(wine1)
```

```python
wine_normal
```

```python
pca = PCA()
pca_values = pca.fit_transform(wine_normal)
```

```python
pca_values
```

```python
pca = PCA(n_components = 13)
pca_values = pca.fit_transform(wine_normal)
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
 pca_values[:,0:1]
```

```python
# plot between PCA1 and PCA2 
x = pca_values[:,0:1]
y = pca_values[:,1:2]
plt.scatter(x,y)
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3']), wine[['Type']]], axis = 1)
finalDf
```

```python
import seaborn as sns
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=finalDf)
```

Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(16,10))
dendrogram=sch.dendrogram(sch.linkage(wine_normal,'complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc = hc.fit_predict(wine_normal)
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

```python
wine['h_clusterid'] = Clusters
wine
```

K - Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,13):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(wine_normal)
    wcss.append(kmeans.inertia_)
    
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,13),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS');
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(3, random_state=42)
clusters_new.fit(wine_normal)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
wine['clusterid_new'] = clusters_new.labels_
wine
```
