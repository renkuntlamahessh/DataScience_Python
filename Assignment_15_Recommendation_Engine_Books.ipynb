```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv("book.csv",encoding='latin-1')
```

```python
books.columns
```

```python
books.columns=['sr_no.','user_id','book_title','book_rating']
```

```python
books.head()
```

```python
#number of unique users in the dataset
len(books.user_id.unique())
```

```python
len(books.book_title.unique())
```

```python
books_matrix = books.pivot_table(index='user_id',
                                         columns='book_title',
                                         values='book_rating').reset_index(drop=True)
```

```python
books_matrix
```

```python
books_matrix.index = books.user_id.unique()
```

```python
books_matrix
```

```python
#Impute those NaNs with 0 values
books_matrix.fillna(0,inplace=True)
```

```python
books_matrix
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(books_matrix.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books.user_id.unique()
user_sim_df.columns = books.user_id.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:125]
```
